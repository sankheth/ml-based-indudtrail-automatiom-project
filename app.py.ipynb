```python
import streamlit as st
import joblib
import numpy as np

model = joblib.load("sensor_model.pkl")
encoder = joblib.load("label_encoder.pkl")

st.title("AI Safety Monitoring System")

temp = st.number_input("Temperature", value=30.0)
gas = st.number_input("Gas Sensor", value=200)
flame = st.selectbox("Flame Detected?", [0,1])

if st.button("Predict"):

    sample = np.array([[temp, gas, flame]])
    pred = model.predict(sample)
    result = encoder.inverse_transform(pred)[0]

    issues = []
    actions = []

    # FIRE CHECK
    if flame == 1:
        issues.append("🔥 Fire Detected")
        actions.append("Evacuate immediately.")
        actions.append("Activate fire alarm.")
        actions.append("Use extinguisher only if safe.")
        actions.append("Call emergency services.")

    # GAS CHECK
    if gas > 600:
        issues.append("🧪 Dangerous Gas Level")
        actions.append("Open doors/windows for ventilation.")
        actions.append("Turn off gas supply if safe.")
        actions.append("Do not switch electrical devices ON/OFF.")
        actions.append("Inspect leakage source.")

    elif gas > 350:
        issues.append("🧪 Moderate Gas Rise")
        actions.append("Increase ventilation.")
        actions.append("Check nearby gas equipment.")

    # TEMP CHECK
    if temp > 50:
        issues.append("🌡 High Temperature")
        actions.append("Inspect overheating equipment.")
        actions.append("Reduce heat source immediately.")
        actions.append("Monitor area for fire risk.")

    elif temp > 40:
        issues.append("🌡 Temperature Rising")
        actions.append("Check ventilation/cooling system.")

    # FINAL OUTPUT
    if not issues:
        st.success("✅ All Parameters Normal")
        st.write("No action required.")

    else:
        st.error("⚠ Abnormal Condition Detected")

        st.subheader("Detected Issues:")
        for i in issues:
            st.write("-", i)

        st.subheader("Recommended Actions:")
        for a in dict.fromkeys(actions):
            st.write("-", a)

        st.caption("AI Model Status: " + result)
```
